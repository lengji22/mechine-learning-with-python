```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('moviedataset/ml-latest/ratings.csv')
df.head()
```

```python
df = df.drop('timestamp',axis='columns')
```

```python
df.head()
```

```python
df1 = pd.read_csv('moviedataset/ml-latest/movies.csv')
df1.head()
```

```python
df1['genres'] = df1.genres.str.split('|')
df1.head()
```

```python
main_df = df1.copy()
```

```python
for index, row in main_df.iterrows():
    for genre in row['genres']:
        main_df.at[index, genre] = 1
        
        
```

```python
main_df.head()
```

```python
main_df.fillna(0,inplace=True)
```

```python
userInput = [
            {'title':'The Butterfly Circus (2009)', 'rating':5},
            {'title':'Zero (2015)', 'rating':3.5},
            {'title':'Grand Slam (1967)', 'rating':2},
            {'title':"Grumpier Old Men (1995)", 'rating':5},
            {'title':'Jumanji (1995)', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
user_movies = pd.merge(main_df,inputMovies,on='title')
```

```python
user_movies
```

```python
user_movies_mat = user_movies.drop('movieId', 1).drop('title', 1).drop('genres', 1)
```

```python
user_movies_mat.head()
```

```python
userProfile = user_movies_mat.transpose().dot(inputMovies['rating'])
userProfile
```

```python
genreTable = main_df.drop('title', 1).drop('genres', 1)
```

```python
genreTable.head()
```

```python
user_recomm = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
```

```python
user_recomm.head()
```

```python
user_recomm.sort_values(ascending=False)
```

```python
df1.loc[df1['movieId'].isin(user_recomm.head(20).keys())]
```
